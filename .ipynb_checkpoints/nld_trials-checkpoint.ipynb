```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(-np.pi, np.pi,256, endpoint = True)
C, S=np.cos(X), np.sin(X)
#Defaults
plt.plot(X, C)
plt.plot(X, S)

plt.show()
```

```python
plt.figure(figsize=(10,6), dpi=80)
plt.plot(X, C, color = "blue", linewidth = 2.5, linestyle = "-", label = "cosine")
plt.plot(X, S, color = "red", linewidth = 2.5, linestyle = "-", label = "sin")

plt.xlim(X.min() * 1.1 , X.max() * 1.1)
plt.ylim(C.min() * 1.1 , C.max() * 1.1)

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1, 0, 1])

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],[r'$-\pi$', r'$-\pi/2$', r'$0$',r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1,0,1],[r'$-1$',r'$0$',r'$+1$'])

ax = plt.gca() # gca stands for get current axes
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

font = {'family':'serif',
        'color' :'darkred',
        'weight':'normal',
        'size'  :'16'
}
plt.title('Sin and Cosine Function', fontdict = font)
plt.legend(loc='upper left')

#Annotating intermediate points

t = 2 * np.pi / 3
plt.plot([t, t], [0, np.cos(t)], color='blue', linewidth=2.5, linestyle='--')
plt.scatter([t, ],[np.cos(t), ], 50, color='blue')

plt.annotate(r'$cos(\frac{2\pi}{3})=-\frac{1}{2}$',
              xy=(t,np.cos(t)),xycoords='data', xytext=(-90, -50),textcoords='offset points', fontsize=16,
              arrowprops=dict(arrowstyle="->",connectionstyle="arc3,rad=.2"))
plt.plot([t, t],[0,np.sin(t)],color='red',linewidth=2.5,linestyle="--")
plt.scatter([t, ],[np.sin(t),],50,color='red')

plt.annotate(r'$sin(\frac{2\pi}{3})=\frac{\sqrt{3}}{2}$',
              xy=(t,np.sin(t)),xycoords='data', xytext=(+10, +30),textcoords='offset points', fontsize=16,
              arrowprops=dict(arrowstyle="->",connectionstyle="arc3,rad=.2"))
plt.show()



```

```python
#plotting quiver plots
x,y = np.meshgrid(np.arange(0.01,2,.25), np.arange(-2,2,.25))
z=-x*np.log(x)

fig, ax = plt.subplots(1)

#v is the y component of arrow vectors
#u is the x component of arrow vectors
q = ax.quiver(x,y,x,z,edgecolor ='k')

plt.show()
```

```python
def my_3d_func(x,y):
    return np.sin(x*y)
```

```python
from mpl_toolkits.mplot3d import Axes3D

X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)
Z = my_3d_func(X, Y)

fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X,Y,Z,rstride=1,cstride=1,cmap=plt.cm.hot)
ax.contourf(X,Y,Z,zdir='z',offset=-2, cmap=plt.cm.hot)
ax.set_zlim(-2,2)
ax = plt.gca()
ax.set_title('3D plot')
plt.show()
```

```python
plt.contourf(X,Y,Z,cmap=plt.cm.hot)
plt.contour(X,Y,Z,colors='black')
ax = plt.gca()
ax.set_title('Contour plot')
plt.show()
```

```python
import seaborn as sns
X=np.arange(-1, 1.25, 0.25)
Y=np.arange(-1, 1.25, 0.25)
X, Y=np.meshgrid(X, Y)
Z = my_3d_func(X,Y)

fig1, ax1= plt.subplots(1,1)
g=sns.heatmap(Z,ax=ax1)
g.set_xticklabels(np.arange(-1,1.25,0.25),rotation=0,fontsize = 8)
g.set_yticklabels(['a','b','c','d','e','f','g','h','i'],rotation = 0,fontsize = 8)
plt.xlabel("Values on X axis")
plt.ylabel("Values on Y axis")
plt.title('Heatmap of my 3D function')
```

```python
# Bifurcation
xa=.2
xb=1.99

C=np.linspace(xa,xb,100)
i=1000
Y= np.ones(len(C))
# for x in range(iter):
#       Y= Y*C*(1-Y)#Y**2 -C #get rid of early transients
fig, ax = plt.subplots(figsize=(8,6))

for x in range(i):
    Y = Y**2 - C
    ax.plot(C,Y,'.',color='g',markersize=.5)
ax.set_axis_off()
plt.title("Simple Bifurcation plot")
plt.show()
```

```python
import numpy as np
import scipy as sp
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
```

```python
'''
Here we are dealing with a system of linear differential equations
dx/dt = ax -b(xy)
dy/dt = -cy + d(xy)

x,y= z
z is a tuple of x,y
'''

def lotkavolterra(t,z,a,b,c,d):
    x,y=z
    return [a*x - b*x*y, -c*y + d*x*y]
```

```python
sol = solve_ivp(lotkavolterra,[0,15],[10,5],method='RK45', args=(1.5, 1, 3, 1), dense_output=True)
```

```python
t = np.linspace(0,15,300)
z=sol.sol(t)
plt.plot(t,z.T)
plt.xlabel('t')
plt.legend(['x','y'],shadow=True)
plt.title('Lotka-Volterra System')

plt.show()
```
