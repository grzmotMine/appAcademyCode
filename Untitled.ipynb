```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0.001, 3, 10000,endpoint = True)
T = -2*X*np.log(0.5*X)
plt.plot(X, T, color = "blue", linewidth = 2.5, linestyle = "-", label = "a=1,b=1 Tumor Growth")
plt.plot(X,T)

ax = plt.gca() # gca stands for get current axes
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

font = {'family':'serif',
        'color' :'darkred',
        'weight':'normal',
        'size'  :'16'
}
plt.title('a=2,b=0.5 Tumor Growth', fontdict = font)

plt.show()
```
